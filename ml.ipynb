```python
from sklearn.datasets import make_blobs
raw_data = make_blobs(n_samples=800, n_features=3, centers=4 ,cluster_std=1.2)
raw_data
import matplotlib.pyplot as plt
```

```python
#plt.scatter(raw_data[0][:,0], raw_data[0][:,1], c=raw_data[1])
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(raw_data[0][:, 0], raw_data[0][:, 1],raw_data[0][:,2], c=raw_data[1])
plt.show()
```

```python

import pandas as pd

import numpy as np

import seaborn

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
from sklearn.cluster import KMeans, SpectralClustering
from sklearn.model_selection import train_test_split
model = SpectralClustering(4)
X1, X2, Y1, Y2 = train_test_split(raw_data[0], raw_data[1], train_size=0.25, shuffle=True)
```

```python
model.fit(X1)
```

```python
model.cluster_centers_
```

```python
model.labels_
```

```python
f = plt.figure()
ax1 = f.add_subplot(projection='3d')
ax2 = f.add_subplot(projection='3d')

ax1.set_title('Our Model')

ax1.scatter(X1[:,0], X1[:,1],X1[:,2],c=model.labels_)

ax2.set_title('Original Data')

ax2.scatter(X1[:,0], X1[:,1], X1[:,2],c=Y1)

plt.show()
```

```python
test_labels = model.predict(X2)

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('Our Model')

ax1.scatter(X2[:,0], X2[:,1],c=test_labels)

ax2.set_title('Original Data')

ax2.scatter(X2[:,0], X2[:,1],c=Y2)
```

```python
train_labels = model.labels_
model.labels_
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X1,train_labels)
```

```python
knn_test_labels = knn.predict(X2)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('Our Model')

ax1.scatter(X2[:,0], X2[:,1],c=knn_test_labels)

ax2.set_title('Original Data')

ax2.scatter(X2[:,0], X2[:,1],c=Y2)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score

print(accuracy_score(Y1,train_labels))
print(accuracy_score(Y2,test_labels))
print(accuracy_score(Y2, knn_test_labels))
```

```python
from sklearn.metrics import classification_report

from sklearn.metrics import confusion_matrix

print(classification_report(Y2, knn_test_labels))

print(confusion_matrix(Y2, knn_test_labels))
```
