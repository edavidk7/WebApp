```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "db/dailyActivity_merged.csv"
df = pd.read_csv(file)
df
```

```python
Ids = list(x for x in df["Id"].unique())
```

```python
dataframes = []
for id in Ids:
    dataframes.append(df[df["Id"]==id])
```

```python
dataframes[0]
```

```python
for frame in dataframes:
    pd.to_datetime(frame["ActivityDate"])
```

```python
subjects = []
for frame in dataframes:
    id = frame["Id"].unique()[0]
    mean_steps = int(sum(frame["TotalSteps"])/len(frame))
    mean_dist = int(sum(frame["TotalDistance"])/len(frame))
    mean_cals = int(sum(frame["Calories"])/len(frame))
    subjects.append([id,mean_steps,mean_dist,mean_cals])
```

```python
print(subjects[:5])
```

```python
import sklearn
from sklearn.preprocessing import MinMaxScaler
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(list(x[1] for x in subjects), list(x[2] for x in subjects),
           list(x[3] for x in subjects))
plt.show()
```

```python
df2 = pd.DataFrame(subjects)
df2.set_index(df2[0], inplace=True)
df2.drop(columns=[0], inplace=True)
df2
```

```python
scaler = MinMaxScaler()
scaler.fit(df2[1].values.reshape(-1,1))
df2[1] = scaler.transform(df2[1].values.reshape(-1, 1))

scaler.fit(df2[2].values.reshape(-1, 1))
df2[2] = scaler.transform(df2[2].values.reshape(-1, 1))

scaler.fit(df2[3].values.reshape(-1, 1))
df2[3] = scaler.transform(df2[3].values.reshape(-1, 1))
df2
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(df2[1],df2[2],df2[3])
plt.show()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=4)
```

```python
y_pred = model.fit_predict(df2[[1,2,3]])
```

```python
y_pred
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection="3d")
ax.scatter(df2[1], df2[2], df2[3], c=y_pred)
plt.show()
```

```python
sse = []
for k in range(1,11):
    km = KMeans(n_clusters=k)
    y_pred = km.fit_predict(df2[[1,2,3]])
    sse.append(km.inertia_)
```

```python
plt.plot(list(x for x in range(1,11)),sse)
```

```python
df2["classes"] = y_pred
df2
```

```python
from sklearn.decomposition import PCA
pca = PCA(0.95)
X_tf = pca.fit_transform(df2[[1,2,3]])
X_tf.shape
```

```python
X_tf
```

```python
plt.scatter(X_tf[:,0],X_tf[:,1])
```

```python
from sklearn.cluster import KMeans, SpectralClustering
model1,model2 = KMeans(n_clusters=4), SpectralClustering(n_clusters=4)
y_pred1, y_pred2 = model1.fit_predict(X_tf), model2.fit_predict(X_tf)
#plt.scatter(X_tf[:,0],X_tf[:,1],c=y_pred)
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))

ax1.set_title('KMenas')

ax1.scatter(X_tf[:,0], X_tf[:,1],c=y_pred1)

ax2.set_title('Spectral')

ax2.scatter(X_tf[:,0], X_tf[:,1],c=y_pred2)
```

```python
df2["classes"] = y_pred2
df2
```

```python
df3 = pd.read_csv("db/minuteSleep_merged.csv")
df3
```

```python
dates = pd.to_datetime(df3["date"])
```

```python
df3["date"] = dates
df3["date"]
```

```python
print(len(df3["Id"].unique()))
```

```python
dif_dict = {"ID": [], "AvgSleepMins": []}
for id in df3["Id"].unique():
    sub_df = df3.loc[df3["Id"] == id]
    deltas = []
    skip = False
    for dates in sub_df["date"].dt.date.unique():
        temp_df = sub_df.loc[sub_df["date"].dt.date == dates]
        temp_df.reset_index(inplace=True)
        D1 = temp_df["date"][0]
        D2 = temp_df["date"][len(temp_df)-1]
        time_d = int(((D2-D1).seconds/60))
        if time_d <= 720:
            deltas.append(time_d)
    time_mean = np.mean(deltas)
    dif_dict["ID"].append(id)
    dif_dict["AvgSleepMins"].append(time_mean)
df4 = pd.DataFrame(dif_dict)
df4
```
