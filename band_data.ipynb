```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_feather("data/dfs/raw_band_data.feather")
df["TIMESTAMP"] = pd.to_datetime(df["TIMESTAMP"], unit='s')
df["TIMESTAMP"] = df["TIMESTAMP"].dt.tz_localize("CET")
df
```

```python
df_ids = pd.read_feather("data/dfs/device.feather")
```

```python
df_ids.rename({"_id":"DEVICE_ID"}, inplace=True, axis=1)
df_ids.drop(["NAME","MANUFACTURER","IDENTIFIER","TYPE","MODEL"], inplace=True, axis=1)
df_ids
```

```python
new_df = pd.merge(df,df_ids, on="DEVICE_ID",how="inner")
new_df.drop(["DEVICE_ID","USER_ID"],inplace=True, axis=1)
new_df
```

```python
df_aliases = pd.read_excel("data/sample_mod.xlsx")
df_aliases.drop(["Časová značka", "Identifikace měření"],inplace=True,axis=1)
df_aliases.rename({"Alias":"ALIAS"},inplace=True, axis=1)
df_aliases
```

```python
merged_all = new_df.copy()
merged_all["ALIAS"] = merged_all["ALIAS"].astype("category")
df_aliases["ALIAS"] = df_aliases["ALIAS"].astype("category")
merged_all = pd.merge(merged_all, df_aliases, on="ALIAS",how="right")
merged_all
```

```python
merged_all["DATETIME"] = merged_all["TIMESTAMP"].dt.tz_localize("utc")
merged_all["DATETIME_CET"] = merged_all["DATETIME"].dt.tz_convert("CET")
merged_all
```

```python
screening_start_1 = pd.to_datetime("2021-05-10 00:00:00+02:00")
screening_end_1 = pd.to_datetime("2021-05-17 00:00:00+02:00")
screening_start_2 = pd.to_datetime("2021-05-17 00:00:00+02:00")
screening_end_2 = pd.to_datetime("2021-05-24 00:00:00+02:00")
alias_col = merged_all["DATETIME_CET"]
filtering_alias = alias_col.map(lambda x: (x >= screening_start_1 and x < screening_end_1) or (
    x >= screening_start_2 and x < screening_end_2))
merged_all = merged_all[filtering_alias]
```

```python
merged_all["WEEK_DAY"] = merged_all["DATETIME_CET"].dt.day_name()
```

```python
schedule = pd.read_csv(
    "data/schedule.csv")
```

```python
merged_all
```

```python
schedule
```

```python
merged_all.reset_index(inplace=True)
merged_all
```

```python
for i, row in merged_all.iterrows():
    subject = False
    if row["WEEK_DAY"] == "Sunday" or row["WEEK_DAY"] == "Saturday":
        merged_all.at[i, 'SCHOOL'] = False
        merged_all.at[i, 'SUBJECT'] = "N/A"
        merged_all.at[i, 'BREAK'] = False
        merged_all.at[i, 'WEEKEND'] = True
    else:
        merged_all.at[i, 'WEEKEND'] = False
        day_schedule = schedule[schedule["day"]==row["WEEK_DAY"]]
        day_schedule.reset_index(inplace=True)
        c_time = row["DATETIME_CET"].time()
        #day_schedule["from"] = pd.Timestamp(day_schedule["from"]).time
        #day_schedule["to"] = pd.Timestamp(day_schedule["to"]).time
        school_start = pd.Timestamp(day_schedule.loc[0].at["from"]).time()
        school_end = pd.Timestamp(
            day_schedule.at[len(day_schedule["to"])-1, "to"]).time()
        for k, row in day_schedule.iterrows():
            subject_start = pd.Timestamp(row["from"]).time()
            subject_end = pd.Timestamp(row["to"]).time()
            if (subject_start <= c_time) and (c_time <= subject_end):
                merged_all.at[i,'SUBJECT'] = row["subject"]
                merged_all.at[i, 'BREAK'] = False
                merged_all.at[i, 'SCHOOL'] = True
                subject = True
                
        if school_start <= c_time and c_time <= school_end and subject==False:
            merged_all.at[i, 'SUBJECT'] = "N/A"
            merged_all.at[i, 'BREAK'] = True
            merged_all.at[i, 'SCHOOL'] = True
        elif subject == False: 
            merged_all.at[i, 'SUBJECT'] = "N/A"
            merged_all.at[i, 'BREAK'] = False
            merged_all.at[i, 'SCHOOL'] = False
```
